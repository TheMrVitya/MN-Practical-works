```python
#4 Створіть масив розміром 5x5, заповніть його випадковими дійсними числами в діапазоні від 0 до 1 та обчисліть суму елементів, які менше середнього значення.

import random
matrix = [[random.uniform(0, 1) for _ in range(5)] for _ in range(5)]
print("Значення матриці:")
for row in matrix:
    print(row)

average_value = sum(sum(row) for row in matrix) / (5 * 5)
sum_below_average = sum(element for row in matrix for element in row if element < average_value)
print(f"Середнє значення: {average_value}")
print(f"Сума елементів, які менше середнього значення: {sum_below_average}")
```

```python
#5 Створіть два одновимірних масиви розміром 5, заповніть їх випадковими цілими числами від 1 до 30 та виконайте на них поелементні операції додавання, віднімання та множення.

import random
array1 = [random.randint(1, 30) for _ in range(5)]
array2 = [random.randint(1, 30) for _ in range(5)]
print("Масив 1:", array1)
print("Масив 2:", array2)
sum_array = [x + y for x, y in zip(array1, array2)]
difference_array = [x - y for x, y in zip(array1, array2)]
product_array = [x * y for x, y in zip(array1, array2)]
print("Сума масивів:", sum_array)
print("Різниця масивів:", difference_array)
print("Добуток масивів:", product_array)
```

```python
#6 Створіть два вектори розміром 6, заповніть їх довільними дійсними числами та знайдіть їх скалярний добуток.

import random
vector1 = [random.uniform(1, 10) for _ in range(6)]
vector2 = [random.uniform(1, 10) for _ in range(6)]
print("Вектор 1:", vector1)
print("Вектор 2:", vector2)
dot_product = sum(x * y for x, y in zip(vector1, vector2))
print("Скалярний добуток:", dot_product)
```

```python
#7 Створіть дві матриці розміром 2x3 та 3x2, заповніть їх випадковими цілими числами в діапазоні від 1 до 15 та перемножте їх між собою.

import random
matrix1 = [[random.randint(1, 15) for _ in range(3)] for _ in range(2)]
matrix2 = [[random.randint(1, 15) for _ in range(2)] for _ in range(3)]
print("Матриця 1:")
for row in matrix1:
    print(row)

print("\nМатриця 2:")
for row in matrix2:
    print(row)

result_matrix = [[sum(x * y for x, y in zip(row1, col2)) for col2 in zip(*matrix2)] for row1 in matrix1]
print("\nРезультат перемноження матриць:")
for row in result_matrix:
    print(row)
```

```python
#8 Створіть матрицю розміром 3x3, заповніть її випадковими цілими числами в діапазоні від 1 до 10 та знайдіть її обернену матрицю.

import numpy as np
import random
matrix = np.array([[random.randint(1, 10) for _ in range(3)] for _ in range(3)], dtype=float)
print("Початкова матриця:")
print(matrix)
try:
    inverse_matrix = np.linalg.inv(matrix)
    print("\nОбернена матриця:")
    print(inverse_matrix)
except np.linalg.LinAlgError:
    print("\nМатриця є сингулярною (немає оберненої матриці).")
```

```python
#9 Створіть матрицю розміром 5x5, заповніть її випадковими дійсними числами в діапазоні від 0 до 1 та транспонуйте її.

import numpy as np
import random
matrix = np.array([[random.uniform(0, 1) for _ in range(5)] for _ in range(5)])
print("Початкова матриця:")
print(matrix)
transposed_matrix = np.transpose(matrix)
print("\nТранспонована матриця:")
print(transposed_matrix)
```

```python
#10 Створіть матрицю розміром 3x4 та вектор розміром 4, заповніть матрицю випадковими цілими числами в діапазоні від -5 до 5, а вектор - випадковими дійсними числами від 0 до 1, і знайдіть суму додатних елементів отриманої матриці.

import numpy as np
import random
matrix = np.array([[random.randint(-5, 5) for _ in range(4)] for _ in range(3)])
print("Матриця:")
print(matrix)
vector = np.array([random.uniform(0, 1) for _ in range(4)])
print("\nВектор:")
print(vector)
positive_sum = np.sum(matrix[matrix > 0])
print("\nСума додатних елементів матриці:", positive_sum)
```

```python
#12 Створіть дві матриці розміром 3x3, заповніть їх випадковими цілими числами в діапазоні від 1 до 10 та знайдіть добуток всіх елементів обох матриць.

import numpy as np
import random
matrix1 = np.array([[random.randint(1, 10) for _ in range(3)] for _ in range(3)])
matrix2 = np.array([[random.randint(1, 10) for _ in range(3)] for _ in range(3)])
print("Матриця 1:")
print(matrix1)
print("\nМатриця 2:")
print(matrix2)

product = np.prod(matrix1) * np.prod(matrix2)
print("\nДобуток всіх елементів обох матриць:", product)
```

```python
#13 Створіть дві матриці розміром 2x2, заповніть їх випадковими цілими числами в діапазоні від 1 до 10 та знайдіть різницю їх елементів.

import numpy as np
import random
matrix1 = np.array([[random.randint(1, 10) for _ in range(2)] for _ in range(2)])
matrix2 = np.array([[random.randint(1, 10) for _ in range(2)] for _ in range(2)])
print("Матриця 1:")
print(matrix1)
print("\nМатриця 2:")
print(matrix2)
difference_matrix = matrix1 - matrix2
print("\nРізниця елементів матриць:")
print(difference_matrix)
```

```python
#14 Створіть матрицю розміром 8x8, заповніть її випадковими цілими числами в діапазоні від 1 до 100 та знайдіть кількість парних елементів матриці.

import numpy as np
import random
matrix = np.array([[random.randint(1, 100) for _ in range(8)] for _ in range(8)])
print("Матриця:")
print(matrix)
count_even_elements = np.sum(matrix % 2 == 0)
print("\nКількість парних елементів матриці:", count_even_elements)
```

```python
#15 Створіть матрицю розміром 4x4, заповніть її випадковими дійсними числами в діапазоні від -1 до 1 та знайдіть вектор-стовпчик, що містить суму кубів елементів кожного рядка матриці.

import numpy as np
import random
matrix = np.array([[random.uniform(-1, 1) for _ in range(4)] for _ in range(4)])
print("Матриця:")
print(matrix)
cubed_sum_vector = np.sum(matrix**3, axis=1, keepdims=True)
print("\nВектор-стовпчик сум кубів елементів кожного рядка:")
print(cubed_sum_vector)
```
